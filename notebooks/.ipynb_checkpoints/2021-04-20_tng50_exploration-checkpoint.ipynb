```python
import numpy as np
from matplotlib import pyplot as plt
import sys

# shouldn't need this if have illustris_python properly in python path! todo: check if fixed upon reload
sys.path.insert(1, '/home/ksf293/external')
import illustris_python as il
```

Load in subhalos for redshift 0 (snapshot 99) of the smallest, lowest-res TNG box, TNG50-4:

```python
basePath = '/scratch/ksf293/gnn-cosmology/data/TNG50-4/output'
snap_num = 99
fields = ['SubhaloMass','SubhaloPos','SubhaloMassType']
subhalos = il.groupcat.loadSubhalos(basePath,snap_num,fields=fields)
```

```python
print(subhalos.keys())
print(subhalos['SubhaloMassType'].shape)
```

```python
halos = il.groupcat.loadHalos(basePath,snap_num)
print(halos.keys())
print(halos['GroupMass'].shape)
```

Plot distribution of subhalo masses (includes all particle types):

```python
mass = subhalos['SubhaloMass'] * 1e10
plt.figure(figsize=(8,6))
bins = np.logspace(np.log10(min(mass)), np.log10(max(mass)), 20)
plt.hist(mass, bins=bins, log=True, histtype='step', lw=2, color='grey')
plt.xscale('log')
plt.xlabel(r'Total mass, M$_\odot$')
plt.ylabel('number')
```

Now divide this into gas, stars, and dm:

```python
mgas, mdm, mstars = subhalos['SubhaloMassType'][:,[0,1,4]].T * 1e10
```

```python
bins = np.logspace(6, 15, 20)
lw = 2
plt.figure(figsize=(8,6))
plt.hist(mgas, bins=bins, log=True, histtype='step', label='Gas', lw=lw)
plt.hist(mstars, bins=bins, log=True, histtype='step', label='Stars', lw=lw)
plt.hist(mdm, bins=bins, log=True, histtype='step', label='DM', lw=lw)
plt.xscale('log')
plt.xlabel(r'Mass, M$_\odot$')
plt.ylabel('number')
plt.legend()
```

## Subhalo positions

SubhaloPos: Spatial position within the periodic box (of the particle with the minium gravitational potential energy). Comoving coordinate. (ckpc/h)

```python
pos = subhalos['SubhaloPos']
```

```python
min(pos[:,0]), max(pos[:,0])
```

```python
zslice = 0, 5000
```

```python
pos_inzslice = pos[np.where((zslice[0] < pos[:,2]) & (pos[:,2] < zslice[1]))]
```

```python
plt.figure(figsize=(6,6))
plt.scatter(pos_inzslice[:,0], pos_inzslice[:,1], s=6, color='blue')
plt.xlabel(r'$x$ ($c\,$kpc/$h$)')
plt.ylabel(r'$y$ ($c\,$kpc/$h$)')
```

GroupFirstSub: Index into the Subhalo table of the first/primary/most massive Subfind group within this FoF group. Note: This value is signed (or should be interpreted as signed)! In this case, a value of -1 indicates that this FoF group has no subhalos.

```python
GroupFirstSub = il.groupcat.loadHalos(basePath,99,fields=['GroupFirstSub'])
```

```python
GroupFirstSub.shape
```

Let's look at the most massive halo:

```python
haloID = 0 
subhaloID = GroupFirstSub[haloID]
#all_fields = il.groupcat.loadSingle(basePath,99,subhaloID=GroupFirstSub[i])
halo_dm = il.snapshot.loadHalo(basePath,snap_num,haloID,'dm')
subhalo_stars = il.snapshot.loadSubhalo(basePath,snap_num,subhaloID,'stars')
subhalo_gas = il.snapshot.loadSubhalo(basePath,snap_num,subhaloID,'gas')
```

```python
halo_dm.keys()
subhalo_stars.keys()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(halo_dm['Coordinates'][:,0], halo_dm['Coordinates'][:,1], s=1, marker='.', color='darkblue', label='Halo DM')
plt.scatter(subhalo_gas['Coordinates'][:,0], subhalo_gas['Coordinates'][:,1], alpha=0.1, s=1, marker='.', color='magenta', label='Subhalo_0 Gas')
plt.scatter(subhalo_stars['Coordinates'][:,0], subhalo_stars['Coordinates'][:,1], s=1, marker='.', color='orange', label='Subhalo_0 Stars')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.xlabel(r'$x$ ($c\,$kpc/$h$)')
plt.ylabel(r'$y$ ($c\,$kpc/$h$)')
```

Now let's pick a smaller one:

```python
haloID = 500 
subhaloID = GroupFirstSub[haloID]
#all_fields = il.groupcat.loadSingle(basePath,99,subhaloID=GroupFirstSub[i])
halo_dm = il.snapshot.loadHalo(basePath,snap_num,haloID,'dm')
subhalo_stars = il.snapshot.loadSubhalo(basePath,snap_num,subhaloID,'stars')
subhalo_gas = il.snapshot.loadSubhalo(basePath,snap_num,subhaloID,'gas')
```
