```python
import numpy as np
from matplotlib import pyplot as plt

from astropy.table import Table
```

```python
import matplotlib
from matplotlib import pylab
%config InlineBackend.figure_format = 'retina'
matplotlib.rcParams['figure.dpi'] = 80
params = {'legend.fontsize': 'xx-large',
          'figure.figsize': (6, 4),
         'axes.labelsize': 'xx-large',
         'axes.titlesize': 'xx-large',
         'xtick.labelsize': 'x-large',
         'ytick.labelsize': 'x-large',
         'figure.facecolor': 'w'}
pylab.rcParams.update(params)
```

```python
sim_name = 'TNG100-1'
fn_halos = f'../data/halo_tables/halos_{sim_name}.fits'
```

```python
tab_halos = Table.read(fn_halos)
```

```python
print(len(tab_halos))
print(tab_halos.columns)
```

```python
np.min(tab_halos['m200m']), np.max(tab_halos['m200m']), np.median(tab_halos['m200m'])
```

```python
plt.hist(tab_halos['m200m'], bins=50)
plt.xlabel('log(Group_M_Mean200 [Msun/h])')
plt.ylabel('number per bin')
```

```python
x_min = np.min(tab_halos['mstellar'][tab_halos['mstellar'] > -np.inf])
bins = np.linspace(x_min, np.max(tab_halos['mstellar']), 50)
plt.hist(tab_halos['mstellar'], bins=bins)
plt.xlabel('log(mstellar [Msun/h])')
plt.ylabel('number per bin')
```

```python
plt.hist(np.log10(tab_halos['r200m']), bins=50)
plt.xlabel('log(Group_R_Mean200)')
plt.ylabel('number per bin')
```

```python
sorted(tab_halos['r200m'])[300:310]
```

```python
plt.figure(figsize=(6,6))
xx = np.linspace(np.min(np.log10(tab_halos['r200m_fof'])), np.max(np.log10(tab_halos['r200m'])))
plt.plot(xx, xx, color='grey')
plt.scatter(np.log10(tab_halos['r200m']), np.log10(tab_halos['r200m_fof']), color='k', alpha=0.2, s=1)
plt.xlabel('log(Group_R_Mean200)')
plt.ylabel('log(m200m_fof)')
```
