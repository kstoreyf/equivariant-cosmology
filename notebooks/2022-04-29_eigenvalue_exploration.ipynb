```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = [0, 0]
x_var = np.array([[0.1, 0], [0,0.1]])
```

```python
n = 1000
x_points = np.random.multivariate_normal(mean, x_var, size=n)
```

```python
eigvals = np.linalg.eigvalsh(x_var)
print(eigvals)
```

```python
plt.figure(figsize=(5,5))
ax = plt.gca()
ax.scatter(x_points[:,0], x_points[:,1], color='k', s=6)
ax.axvline(0, lw=0.5)
ax.axhline(0, lw=0.5)
ax.set_aspect('equal')
circle = plt.Circle((0, 0), 1, color='r', fill=False, lw=2)
ax.add_patch(circle)
```

```python
x_var_empirical = np.mean([np.outer(x_points[i], x_points[i]) for i in range(len(x_points))], axis=0)
```

```python
print(x_var_empirical)
print(np.linalg.eigvalsh(x_var_empirical))
```

```python
def get_circle(r, n):
    x = []
    y = []
    for i in range(n):
        angle = np.random.uniform(0,1)*(np.pi*2)
        x.append(r*np.cos(angle));
        y.append(r*np.sin(angle));
    return np.array([x, y]).T
```

```python
x_points = get_circle(0.95, n)
```

```python
plt.figure(figsize=(5,5))
ax = plt.gca()
ax.scatter(x_points[:,0], x_points[:,1], color='k', s=6)
ax.axvline(0, lw=0.5)
ax.axhline(0, lw=0.5)
ax.set_aspect('equal')

circle = plt.Circle((0, 0), 1, color='r', fill=False, lw=2)
ax.add_patch(circle)
```

```python
x_var = np.mean([np.outer(x_points[i], x_points[i]) for i in range(len(x_points))], axis=0)
```

```python
print(x_var)
print(np.linalg.eigvalsh(x_var))
```
